```python
import pandas as pd
import sqlite3
```

```python
str_path = r"..\db\db_itens_revision.db"
str_sql = f"SELECT COD_ITEM, SM, NEW_SHORT_TEXT, NEW_LONG_TEXT, OLD_TEXT, DT FROM ITENS WHERE STATUS != '' AND DT >= '2024.08.23'"

conn = sqlite3.connect(str_path)
df = pd.read_sql(str_sql, conn)

conn.close()
print(f"Total: {len(df)}")
df.head(3)
```

```python
df["N_CARAC"] = df["NEW_SHORT_TEXT"].apply(len)
df
```
